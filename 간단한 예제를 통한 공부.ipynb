### 초등학생 수학 연습 프로그램
단, 나눗셈을 하여 나머지는 버림 하여라

```python
import random
import math
def cal(num1,num2,op):
    if op=="+":
        result=num1+num2
    elif op=="-":
        result=num1-num2
    elif op=="*":
        result=num1*num2
    elif op=="/":
        result=num1/num2
        result=math.floor(result)
    return result

num1=0
num2=0
op=""
userinput=0
result=0
oplist=['+','-','*','/']
for i in range(0,5,1):
    num1=random.randrange(0,10)
    num2=random.randrange(1,10)
    op=random.choice(oplist)
    print(num1,op,num2,'=',end="")
    userinput=int(input())
    result=cal(num1,num2,op)
    if userinput==result:
        print("정답")
    else:
        print("오답")          
```

### 초등학생 수학 연습 자동 채점 프로그램
- 학생이 푼 문제를 파일로 작성한다.
- 학생이 푼 문제 파일을 읽어 자동채점하고 결과를 파일로 작성한다

```python
import random
import math
oplist=['+','-','*','/']
outf=open("C:/0.ITstudy/6.Python/test1.txt","w")
name=input("이름: ")
strline="이름"+name+"\n"
outf.writelines(strline)
for i in range(0,5,1):
    num1=random.randrange(1,10)
    num2=random.randrange(1,10)
    op=random.choice(oplist)
    print(num1,op,num2,"=",end="")
    userinput=int(input())
    
    strline=str(num1)+" "+op+" "+str(num2)+" "+"="+" "+str(userinput)
    outf.writelines(strline+"\n")
outf.close()
```

```python
import random
import math
def cal(num1,num2,op):
    if op=="+":
        result=num1+num2
    elif op=="-":
        result=num1-num2
    elif op=="*":
        result=num1*num2
    elif op=="/":
        result=num1/num2
    return result
inf=open("C:/0.ITstudy/6.Python/test1.txt","r")
outf=open("C:/0.ITstudy/6.Python/test1_A.txt","w")
name=inf.readline()
outf.writelines(name)
for i in range(0,5,1):
    strline=inf.readline()
    temp=strline.split(" ")
    num1=temp[0]
    num2=temp[2]
    op=temp[1]
    userinput=temp[4]
    result=cal(int(num1),int(num2),op)
    if int(userinput)==result:
        strline=num1+" "+" "+op+" "+num2
        outf.writelines(strline)
        outf.writelines("정답"+"\n")
    else:
        strline=num1+" "+" "+op+" "+num2
        outf.writelines(strline)
        outf.writelines("오답"+"\n")
inf.close()
outf.close()
```
